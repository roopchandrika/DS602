## When there is a missing value in dep_time then this is an indication of canceled flight. Find the number of cancelled flights for each (carrier, origin) combination.

```python
flights['dep_time'].isna().count() # count the missing values in dep_time column
```

```python
cf = flights[flights['dep_time'].isna()]
```

```python
cc = cf.groupby(['carrier', 'origin']).size() #groupby carrier and origin and find out the size of cancelled flights
cc
```

## Which carrier has the worst dep_delays?

```python
delay_carrier = flights.groupby('carrier')['dep_delay'].mean() # get the average of grouped carrier and dep_delay
delay_carrier
```

```python
wc = delay_carrier.idxmax() # get the index of carrier which has worst dep_delays
wc
```

## Which plane (tailnum) has the worst on-time record?

```python
avg_tailnum = flights.groupby('tailnum')['dep_delay'].mean() # get the average of grouped tailnum and dep_delay
avg_tailnum
```

```python
twc = avg_tailnum.idxmax() # get the index of tailnum which has worst dep_delays
twc
```

## For each plane, count the number of flights before the first delay of greater than 1 hour.

```python
# sort values by tailnum and time hour
flights = flights.sort_values(by=['tailnum', 'time_hour'])

# Create a new column which stores the flight greater than 60
flights['del_gt_1_hour'] = flights['dep_delay'] > 60

# Group the DataFrame by 'tailnum' and calculate the cumulative sum
flights['cd_count'] = flights.groupby('tailnum')['del_gt_1_hour'].cumsum()

# Filter the DataFrame to include only flights before the first delay
flights_bfd = flights[flights['cd_count'] == 0]

# count the number of flights before the first delay
flight_cbp = flights_bfd['tailnum'].value_counts()

# Print the results
print(flight_cbp)
```
